```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.matrix([[0.2,0.6,0.9,0.3],
               [0.8,0.7,0.3,0.5]])
plt.plot(data, marker='o')
plt.xticks([0,1],[r'$A_1$',r'$A_2$'])
plt.title(',  4 ', 
          fontsize=14, 
          fontname='Times New Roman')
plt.xlabel('  A', 
          fontsize=14, 
          fontname='Times New Roman')
plt.ylabel(' ', 
          fontsize=14, 
          fontname='Times New Roman')
plt.legend([r'$W_1$',
            r'$W_2$',
            r'$W_3$',
            r'$W_4$'],
           loc=0)
plt.grid(True)
plt.show()
```
