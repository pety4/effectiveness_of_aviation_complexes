```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.matrix([[0.8,0.7,0.3,0.5],
               [0.2,0.6,0.9,0.3]])
#data=np.matrix([[0.6,0.2,0.3,0.9],
#               [0.7,0.8,0.5,0.3]])
plt.plot(data, marker='o')
plt.xticks([0,1],[r'$A_1$',r'$A_2$'])
plt.title('Прямые, образованные 4 уравнениями', 
          fontsize=14, 
          fontname='Times New Roman')
plt.xlabel('Стратегии игрока A', 
          fontsize=14, 
          fontname='Times New Roman')
plt.ylabel('Значение игры', 
          fontsize=14, 
          fontname='Times New Roman')
plt.legend([r'$W_1$',
            r'$W_2$',
            r'$W_3$',
            r'$W_4$'],
           loc=0)
plt.grid(True)
plt.show()
```

```python
#координаты точек пересечения прямых
x=[0, 0.25, 0.75, 1]
y=[0.3, 0.45, 0.35, 0.2]
plt.plot(x,y, marker='o')
plt.grid(True)
plt.title('Нижняя огибающая', 
          fontsize=14, 
          fontname='Times New Roman')
plt.xlabel(r'p', 
          fontsize=14, 
          fontname='Times New Roman')
plt.ylabel(r'W', 
          fontsize=14, 
          fontname='Times New Roman')
plt.show()
```
