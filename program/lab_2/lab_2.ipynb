```python
import numpy as np
import pandas as pd
from IPython.display import Markdown
from math import factorial
import matplotlib.pyplot as plt
```

```python
lambda_var=16
mu=5
n=4
alpha=lambda_var/mu
```

```python
def p_0_count(alpha,n):
    p_0=0.0
    for i in range(n+1):
        p_0+=pow(alpha,i)/factorial(i)
    p_0=1/p_0
    return p_0
```

```python
p_0=np.round(p_0_count(alpha,n),4)
P_otk=np.round(pow(alpha,n)*p_0/factorial(n),4)
q=np.round(1-P_otk,4)
A=np.round(lambda_var*q,4)
N=np.round(alpha*(1-P_otk),4)
Markdown(rf"""
$\alpha={alpha}\\$
$p_0={p_0}\\$
Вероятность преодоления системы ПВО:$\\$
$P_о={P_otk}\\$
Относительная пропускная способность системы ПВО:$\\$
$q={q}\\$
Абсолютная пропускная способность системы ПВО$\\$
$A={A}\\$
Степень загрузки системы ПВО:$\\$
$N={N}\\$
""")
```

```python
n_array=range(n+1)
P_otk=[0]*(n+1)
for i in n_array:
    P_otk[i]=pow(alpha,i)*p_0_count(alpha,i)/factorial(i)
plt.plot(n_array, P_otk)
plt.xlabel('n') #Подпись для оси х
plt.ylabel('$P_{ОТК}$') #Подпись для оси y
plt.title(rf'$\alpha={alpha}$') #Название
plt.show()
```
